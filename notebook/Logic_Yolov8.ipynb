```python
from google.colab import drive
import os
import yaml

# 1. Sambungkan ke Google Drive
drive.mount('/content/drive')

# 2. COPY Data dari Drive ke Colab (Biar Training Ngebut üöÄ)
# Asumsi: Folder kamu ada di 'MyDrive/PCGK'
drive_path = "/content/drive/MyDrive/PCGK"
colab_path = "/content/dataset_pothole"

print(f"‚è≥ Sedang menyalin data dari {drive_path} ke {colab_path}...")
# Pastikan folder tujuan ada
!mkdir -p "{colab_path}"
# Kita pakai perintah Linux 'cp' biar cepat, menyalin isi folder, bukan folder itu sendiri
!cp -r "{drive_path}"/* "{colab_path}"
print("‚úÖ Selesai menyalin! Data siap dipakai.")

# --- Diagnosa: Periksa struktur folder setelah menyalin ---
print(f"\nüîç Memeriksa struktur folder di {colab_path}:")
!ls -R "{colab_path}"

print(f"\nüîç Memeriksa struktur folder di Google Drive ({drive_path}):")
!ls -R "{drive_path}"
# ---------------------------------------------------------

# 3. MEMBUAT FILE data.yaml OTOMATIS
# Ini PENTING biar YOLO tau lokasi folder train/valid yang baru di-copy
# Pastikan di dalam folder PCGK kamu ada folder 'train' dan 'valid' ya
yaml_content = f"""
path: {colab_path}/dataset_pothole  # Corrected path to the nested dataset folder
train: train/images
val: valid/images

nc: 1
names: ['pothole']
"""

# Simpan file data.yaml yang baru
yaml_file = f"{colab_path}/data.yaml"
with open(yaml_file, 'w') as f:
    f.write(yaml_content)

print("‚úÖ File konfigurasi data.yaml berhasil dibuat!")

# 4. INSTALL & TRAINING
# Install dulu
!pip install ultralytics

from ultralytics import YOLO

# Load model
model = YOLO('yolov8n.pt')

# Training (Patience=15 biar kalau udah pinter dia stop sendiri)
print("üöÄ MULAI TRAINING...")
model.train(data=yaml_file, epochs=50, imgsz=640, patience=15)
```
